```python
df=pd.read_csv(r"C:\Users\Admin\Desktop\winequality-white (1).csv")
df.head()
```

```python
import keras
import tensorflow as tf
import numpy as np
from keras.layers import  Dense, Flatten , Activation ,Dropout, InputLayer ,LSTM, Conv1D, MaxPooling1D
from keras.models  import Sequential
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import normalize
from sklearn.metrics import classification_report
```

```python
data=np.array(df)
data.shape
(x_t1,x_t2,y_t1,y_t2)=train_test_split(normalize(data[:,0:11]),data[:,11],test_size=0.15)
#shuff=np.random.permutation(data.shape[0])

#x_t1=data[shuff[:4500],0:data.shape[1]-1].reshape(4500,11)
#y_t1=data[shuff[:4500],data.shape[1]-1].reshape(4500,1)
#x_t2=data[shuff[4500:],0:data.shape[1]-1].reshape(398,11)
#y_t2=data[shuff[4500:],data.shape[1]-1].reshape(398,1)
#print(y_t1)

#one_hot
#y_t1=keras.utils.to_categorical(y_t1,num_classes=10)
#y_t2=keras.utils.to_categorical(y_t2,num_classes=10)
y_t2
```

```python
model = Sequential()

model.add(Dense(128,activation="relu",input_dim=11))
model.add(Dense(256,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(128,activation="relu"))
model.add(Dense(10,activation="softmax"))
```

```python
opt=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

model.compile(optimizer=opt, 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_t1, y_t1, epochs=100,batch_size=64,validation_data=(x_t2,y_t2),shuffle=True)
```

```python
model.evaluate(x_t2,y_t2)
```

```python
pre=model.predict(x_t2)
pre=np.argmax(pre,axis=1)
pre
```

```python
print(classification_report(y_t2,pre))
```
